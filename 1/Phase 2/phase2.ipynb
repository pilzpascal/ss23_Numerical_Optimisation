## Task 1. Newton method with(out) Hessian modification

### Newton Method with Hessian modification

```python
from NM_w_hessian_mod import function_1, function_2, eig_val_mod, add_id, no_mod, newtons_method_hessian_mod
from approx_gradient import approximate_gradient, approximate_hessian

start_points = {'function_1': np.array([[1.2, 1.2], [-1.2, 1], [0.2, 0.8]]),
                'function_2': np.array([[-0.2, 1.2], [3.8, 0.1], [1.9, 0.6]])}
sol_points = {'function_1': np.array([(1., 1.), (1., 1.), (1., 1.)]),
              'function_2': np.array([(4., 0.), (4., 0.), (4., 0.)])}
functions = {'function_1': function_1(),
             'function_2': function_2()}

def perform_nm(func, start_points, sol_points, method):

        nm_dicts = []
        for i, x_0 in enumerate(start_points):
            
            for grad_method, is_approx in [('symbolic', False), ('numeric', True)]:

                f, grad_f, hess_f = func
                if grad_method == 'numeric':
                    grad_f = lambda x: approximate_gradient(f, x)
                    #hess_f = lambda x: approximate_hessian(f, x)

                x_fin, p_fin, a_fin, n_iter = newtons_method_hessian_mod(
                    x_0, f, grad_f, hess_f, stop_crit=1e-6, print_interval=False, hess_modifier=method)
                result = x_fin[-1]

                nm_dicts.append({
                           'Gradient': 'Actual Gradient' if not is_approx else 'Approx. Gradient',
                           '$$x_0$$': r'$$\begin{pmatrix}' + r' \\ '.join([str(round(d, 2)) for d in x_0]) + r'\end{pmatrix}$$',
                           '$$x_k$$': r'$$\begin{pmatrix}' + r' \\ '.join([str(round(d, 2)) for d in result]) + r'\end{pmatrix}$$', 
                           '$$\Vert \nabla f(x_k)\ \Vert$$': np.array(np.linalg.norm(grad_f(result))), 
                           '$$\Vert x^* - x_k \Vert$$': np.array(math.dist(result, sol_points[i])), 
                           'Iterations': np.array(n_iter)
                        })
        display(pd.DataFrame(nm_dicts))
```

## Task 4. QN methods

```python
from qn.optimization import bfgs_method, sr1_method
import math
from qn.problems import rosenbrock, rosenbrock_grad, rosenbrock_starting_points, rosenbrock_solutions
from qn.problems import function_2, function_2_grad, function_2_starting_points, function_2_solutions
```

```python
def perform_QN(method, f, grad_orig, starting_points, sol_points, max_iter=100, epsilon=1e-6):
    qn_dicts = []
    for i, sp in enumerate(starting_points):
        x0 = np.array(sp)
        
        grad_approx = lambda x: approximate_gradient(f, x)
        
        for grad, is_approx in [(grad_orig, False), (grad_approx, True)]:
            x_opt, gnorm, num_iters = method(f, grad, x0, max_iter, epsilon)

            qn_dicts.append({
                       'Gradient': 'Actual Gradient' if not is_approx else 'Approx. Gradient',
                       '$$x_0$$': r'$$\begin{pmatrix}' + r' \\ '.join([str(round(d, 2)) for d in x0]) + r'\end{pmatrix}$$',
                       '$$x_k$$': r'$$\begin{pmatrix}' + r' \\ '.join([str(round(d, 2)) for d in x_opt]) + r'\end{pmatrix}$$', 
                       '$$\Vert \nabla f(x_k)\ \Vert$$': np.array(gnorm), 
                       '$$\Vert x^* - x_k \Vert$$': np.array(math.dist(x_opt, sol_points[i])), 
                       'Iterations': np.array(num_iters)
                    })
    display(pd.DataFrame(qn_dicts))
            
```
