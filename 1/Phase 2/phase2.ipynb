## Task 1. Newton method with(out) Hessian modification

### Newton Method with Hessian modification

```python
from NM_w_hessian_mod import function_1, function_2, eig_val_mod, add_id, no_mod, newtons_method_hessian_mod
from approx_gradient import approximate_gradient, approximate_hessian
import time

start_points = {'function_1': np.array([[1.2, 1.2], [-1.2, 1], [0.2, 0.8]]),
                'function_2': np.array([[-0.2, 1.2], [3.8, 0.1], [1.9, 0.6]])}
sol_points = {'function_1': np.array([(1., 1.), (1., 1.), (1., 1.)]),
              'function_2': np.array([(0., 1.), (4., 0.), (4., 0.)])}
functions = {'function_1': function_1(),
             'function_2': function_2()}

def perform_nm(func, start, sol, method, print_dataframe=True, symbolic_only=False, numeric_only=False, count_time=False):
    
    nm_dicts = []
    grad_methods = [('symbolic', False), ('numeric', True)]
    if symbolic_only:
        grad_methods = grad_methods[:1]
        
    if numeric_only:
        grad_methods = grad_methods[1:]
        
    for i, x_0 in enumerate(start):
        
        for grad_method, is_approx in grad_methods: 

            f, grad_f, hess_f = func
            if grad_method == 'numeric':
                grad_f = lambda x: approximate_gradient(f, x)
                hess_f = lambda x: approximate_hessian(f, x)
            
            start_time = time.perf_counter_ns()   
            x_fin, p_fin, a_fin, n_iter = newtons_method_hessian_mod(
                x_0, f, grad_f, hess_f, stop_crit=1e-6, print_interval=False, hess_modifier=method)
            end_time = time.perf_counter_ns()
            result = x_fin[-1]
            
            nm_dicts.append({
                           'Gradient': 'Actual Gradient' if not is_approx else 'Approx. Gradient',
                           '$$x_0$$': r'$$\begin{pmatrix}' + r' \\ '.join([str(round(d, 4)) for d in x_0]) + r'\end{pmatrix}$$',
                           '$$x_k$$': r'$$\begin{pmatrix}' + r' \\ '.join([str(round(d, 4)) for d in result]) + r'\end{pmatrix}$$',
                           '$$x*$$': r'$$\begin{pmatrix}' + r' \\ '.join([str(round(float(d), 3)) for d in sol[i]]) + r'\end{pmatrix}$$',
                           '$$\Vert \nabla f(x_k)\ \Vert$$': np.array(np.linalg.norm(grad_f(result))), 
                           '$$\Vert x^* - x_k \Vert$$': np.array(math.dist(result, sol[i])),
                           'Iterations': np.array(n_iter)
                           
                        })
            if count_time:
                nm_dicts[-1]['Execution time'] = f'{str(end_time - start_time)}ns'
                
    if print_dataframe:
        display(pd.DataFrame(nm_dicts))
```

## Task 5. Derivatives approximation

```python
from approx_gradient import rosenbrock_function,\
                            rosenbrock_function_g,\
                            rosenbrock_function_h,\
                            function1, function1_g,\
                            function1_h,\
                            approximate_gradient, approximate_hessian

def approximate(f: callable, rf: callable, af: callable, sps: np.ndarray, names: np.ndarray) -> pd.DataFrame:
    """
    Creates a pandas DataFrame for visualization of the data

    Parameters
    ----------
    f : callable
        Function `f(x)`
    rf : callable
        Gradient Function of `f(x)`
    af : callable
        Approximation function 
    sps : np.ndarray
        List of starting points
    names : np.ndarray
        Names for the first two rows in pandas output

    Returns
    -------
    df_dicts
    """
    
    df_dicts = []
    for sp in sps:
        real = rf(sp)
        appr = af(f, sp)
        
        real_r = np.round(real, 2)
        appr_r = np.round(appr, 2)
        if len(appr.shape) < 2:
            real_r = np.round([real], 2).reshape((len(real), 1))
            appr_r = np.round([appr], 2).reshape((len(appr), 1))
        
        real_str = []
        appr_str = []
        for rd1, ad1 in zip(real_r, appr_r):
            real_str.append(r' \ '.join([str(round(rd2, 4)) for rd2 in rd1]))
            appr_str.append(r' \ '.join([str(round(ad2, 4)) for ad2 in ad1]))
        
        df_dicts.append({
                        'Starting Point': r'$$\begin{pmatrix}' + r' \\ '.join([str(d) for d in sp]) + r'\end{pmatrix}$$',
                        f'{names[0]}': r'$$\begin{pmatrix}' + r' \\ '.join([str(d) for d in real_str]) + r'\end{pmatrix}$$',
                        f'{names[1]}': r'$$\begin{pmatrix}' + r' \\ '.join([str(d) for d in appr_str]) + r'\end{pmatrix}$$', 
                        f'{names[2]}': np.linalg.norm(real - appr),
                    })
    return df_dicts
    
rosenbrock_function_sp = np.array([[1.2, 1.2], [-1.2, 1], [0.2, 0.8]])
function1_sp = np.array([[-0.2, 1.2], [3.8, 0.1], [1.9, 0.6]])
```
